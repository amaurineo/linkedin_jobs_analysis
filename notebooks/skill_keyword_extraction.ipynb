```python
import re
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer  # Portuguese stemmer

# Download Portuguese stopwords
import nltk
nltk.download('stopwords')
nltk.download('rslp')

stop_words = stopwords.words('portuguese')
stemmer = RSLPStemmer()

def preprocess(text):
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-ZÀ-ú]', ' ', text)
    # Lowercase and split into tokens
    words = text.lower().split()
    # Remove stopwords and stem
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Example job descriptions list
job_descriptions = pd.read_csv('../data/raw/jobs_data.csv')['job_description'].to_list()

# Preprocess all descriptions
cleaned_docs = [preprocess(doc) for doc in job_descriptions]

# Create TF-IDF matrix
vectorizer = TfidfVectorizer(
    ngram_range=(1, 3),  # Include 1-word, 2-word, and 3-word phrases
    max_features=1000,
    stop_words=stop_words
)
tfidf_matrix = vectorizer.fit_transform(cleaned_docs)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# Train LDA model
n_topics = 3  # Adjust based on your dataset
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(tfidf_matrix)
```

```python
def print_topics(model, vectorizer, n_words=10):
    feature_names = vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(model.components_):
        top_terms = [feature_names[i] for i in topic.argsort()[:-n_words - 1:-1]]
        print(f"Topic {topic_idx}: {', '.join(top_terms)}")

print_topics(lda, vectorizer)
```
